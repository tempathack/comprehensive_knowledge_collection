## 1. What is Data Governance and Why It Matters

### Definition
**Data Governance** is the overall management of data availability, usability, integrity, and security within an organization. It encompasses the policies, standards, processes, and technologies that ensure data is properly managed throughout its lifecycle.

### Core Components

| Component | Description |
|-----------|-------------|
| **Data Stewardship** | Assigning ownership and accountability for data assets |
| **Data Quality** | Ensuring accuracy, completeness, consistency, and timeliness |
| **Data Policies** | Rules governing data creation, storage, usage, and disposal |
| **Data Standards** | Consistent formats, naming conventions, and definitions |
| **Data Architecture** | Technical infrastructure supporting governance |

## 2. Data Cataloging and Metadata Management

### What is a Data Catalog?
A **Data Catalog** is an organized inventory of data assets that uses metadata to help organizations find, understand, and manage their data. It serves as a "search engine" for enterprise data.

### Types of Metadata

| Type | Description | Examples |
|------|-------------|----------|
| **Technical Metadata** | Structure and format of data | Schema, data types, file formats |
| **Business Metadata** | Business context and meaning | Definitions, owners, usage policies |
| **Operational Metadata** | Runtime and process information | Load times, access logs, freshness |
| **Social Metadata** | User-generated annotations | Tags, ratings, comments |

## 3. Data Lineage Tracking

### What is Data Lineage?
**Data Lineage** tracks the origin, movement, and transformation of data throughout its lifecycle. It answers: "Where did this data come from, how was it transformed, and where does it go?"

### Importance of Data Lineage

| Use Case | Description |
|----------|-------------|
| **Impact Analysis** | Understand effects of changes to source data |
| **Root Cause Analysis** | Trace data quality issues to their source |
| **Compliance** | Prove data handling for regulatory requirements |
| **Data Discovery** | Understand how data assets are related |
| **Trust & Transparency** | Build confidence in data for decision making |

````python
# Example: Simple Data Lineage Graph Representation

class DataLineageGraph:
    """Simple representation of data lineage as a directed graph."""
    
    def __init__(self):
        self.nodes = {}  # asset_id -> asset_info
        self.edges = []  # list of (source, target, transformation)
    
    def add_asset(self, asset_id: str, asset_type: str, name: str):
        self.nodes[asset_id] = {
            "type": asset_type,
            "name": name
        }
    
    def add_lineage(self, source_id: str, target_id: str, transformation: str = None):
        self.edges.append({
            "source": source_id,
            "target": target_id,
            "transformation": transformation
        })
    
    def get_upstream(self, asset_id: str) -> list:
        """Get all upstream dependencies."""
        return [e["source"] for e in self.edges if e["target"] == asset_id]
    
    def get_downstream(self, asset_id: str) -> list:
        """Get all downstream dependents."""
        return [e["target"] for e in self.edges if e["source"] == asset_id]
    
    def impact_analysis(self, asset_id: str, visited=None) -> set:
        """Recursively find all downstream assets that would be impacted."""
        if visited is None:
            visited = set()
        
        downstream = self.get_downstream(asset_id)
        for d in downstream:
            if d not in visited:
                visited.add(d)
                self.impact_analysis(d, visited)
        return visited


# Build example lineage
lineage = DataLineageGraph()

# Add data assets
lineage.add_asset("crm_customers", "source", "CRM Customer Table")
lineage.add_asset("pos_transactions", "source", "POS Transactions")
lineage.add_asset("customer_transactions", "derived", "Customer Transactions")
lineage.add_asset("daily_sales", "report", "Daily Sales Report")
lineage.add_asset("customer_360", "derived", "Customer 360 View")

# Add lineage relationships
lineage.add_lineage("crm_customers", "customer_transactions", "JOIN on customer_id")
lineage.add_lineage("pos_transactions", "customer_transactions", "JOIN on customer_id")
lineage.add_lineage("customer_transactions", "daily_sales", "Aggregate by date")
lineage.add_lineage("customer_transactions", "customer_360", "Aggregate by customer")

# Demonstrate impact analysis
print("Impact Analysis: If 'crm_customers' changes, what is affected?")
impacted = lineage.impact_analysis("crm_customers")
for asset_id in impacted:
    print(f"  â†’ {lineage.nodes[asset_id]['name']}")
````

````python
# Example: Role-Based Access Control (RBAC) Implementation

from enum import Enum
from typing import Set

class Permission(Enum):
    READ = "read"
    WRITE = "write"
    DELETE = "delete"
    ADMIN = "admin"
    EXPORT = "export"

class DataClassification(Enum):
    PUBLIC = "public"
    INTERNAL = "internal"
    CONFIDENTIAL = "confidential"
    RESTRICTED = "restricted"

class Role:
    def __init__(self, name: str, permissions: Set[Permission], max_classification: DataClassification):
        self.name = name
        self.permissions = permissions
        self.max_classification = max_classification

# Define roles
ROLES = {
    "viewer": Role("viewer", {Permission.READ}, DataClassification.INTERNAL),
    "analyst": Role("analyst", {Permission.READ, Permission.EXPORT}, DataClassification.CONFIDENTIAL),
    "engineer": Role("engineer", {Permission.READ, Permission.WRITE}, DataClassification.CONFIDENTIAL),
    "admin": Role("admin", {Permission.READ, Permission.WRITE, Permission.DELETE, Permission.ADMIN, Permission.EXPORT}, 
                  DataClassification.RESTRICTED),
}

CLASSIFICATION_HIERARCHY = {
    DataClassification.PUBLIC: 0,
    DataClassification.INTERNAL: 1,
    DataClassification.CONFIDENTIAL: 2,
    DataClassification.RESTRICTED: 3,
}

def check_access(user_role: str, requested_permission: Permission, 
                 data_classification: DataClassification) -> bool:
    """Check if a user with given role can perform action on data."""
    role = ROLES.get(user_role)
    if not role:
        return False
    
    # Check permission
    if requested_permission not in role.permissions:
        return False
    
    # Check classification level
    if CLASSIFICATION_HIERARCHY[data_classification] > CLASSIFICATION_HIERARCHY[role.max_classification]:
        return False
    
    return True

# Test access control
test_cases = [
    ("viewer", Permission.READ, DataClassification.PUBLIC),
    ("viewer", Permission.WRITE, DataClassification.PUBLIC),
    ("analyst", Permission.EXPORT, DataClassification.CONFIDENTIAL),
    ("analyst", Permission.EXPORT, DataClassification.RESTRICTED),
    ("admin", Permission.DELETE, DataClassification.RESTRICTED),
]

print("Access Control Test Results:")
print("-" * 60)
for role, perm, classification in test_cases:
    allowed = check_access(role, perm, classification)
    status = "âœ… ALLOWED" if allowed else "âŒ DENIED"
    print(f"{role:10} | {perm.value:8} | {classification.value:12} | {status}")
````

````python
# Example: Data Subject Request (DSR) Handler for GDPR

from dataclasses import dataclass
from datetime import datetime, timedelta
from typing import Optional, List, Dict
import json

@dataclass
class DataSubjectRequest:
    request_id: str
    request_type: str  # access, erasure, rectification, portability
    subject_id: str
    subject_email: str
    created_at: datetime
    deadline: datetime  # GDPR requires response within 30 days
    status: str  # pending, in_progress, completed, rejected
    notes: Optional[str] = None

class DSRHandler:
    """Handle GDPR Data Subject Requests."""
    
    RESPONSE_DEADLINE_DAYS = 30  # GDPR requirement
    
    def __init__(self):
        self.requests: Dict[str, DataSubjectRequest] = {}
        self.data_sources = ["crm", "analytics", "marketing", "support"]
    
    def create_request(self, request_type: str, subject_id: str, 
                       subject_email: str) -> DataSubjectRequest:
        """Create a new DSR."""
        request_id = f"DSR-{datetime.now().strftime('%Y%m%d')}-{len(self.requests)+1:04d}"
        created_at = datetime.now()
        deadline = created_at + timedelta(days=self.RESPONSE_DEADLINE_DAYS)
        
        dsr = DataSubjectRequest(
            request_id=request_id,
            request_type=request_type,
            subject_id=subject_id,
            subject_email=subject_email,
            created_at=created_at,
            deadline=deadline,
            status="pending"
        )
        self.requests[request_id] = dsr
        return dsr
    
    def handle_access_request(self, request_id: str) -> Dict:
        """Handle Right to Access - export all data for subject."""
        dsr = self.requests.get(request_id)
        if not dsr:
            raise ValueError(f"Request {request_id} not found")
        
        # Simulate collecting data from all sources
        collected_data = {
            "request_id": request_id,
            "subject_id": dsr.subject_id,
            "generated_at": datetime.now().isoformat(),
            "data_by_source": {}
        }
        
        for source in self.data_sources:
            # In reality, query each source for subject's data
            collected_data["data_by_source"][source] = {
                "status": "collected",
                "record_count": 10  # placeholder
            }
        
        dsr.status = "completed"
        return collected_data
    
    def handle_erasure_request(self, request_id: str) -> Dict:
        """Handle Right to Erasure (Right to be Forgotten)."""
        dsr = self.requests.get(request_id)
        if not dsr:
            raise ValueError(f"Request {request_id} not found")
        
        deletion_results = {
            "request_id": request_id,
            "subject_id": dsr.subject_id,
            "deleted_at": datetime.now().isoformat(),
            "results_by_source": {}
        }
        
        for source in self.data_sources:
            # In reality, delete from each source
            deletion_results["results_by_source"][source] = {
                "status": "deleted",
                "records_deleted": 10
            }
        
        dsr.status = "completed"
        return deletion_results


# Demonstrate DSR handling
handler = DSRHandler()

# Create an access request
access_req = handler.create_request(
    request_type="access",
    subject_id="USER-12345",
    subject_email="user@example.com"
)

print(f"Created DSR: {access_req.request_id}")
print(f"Type: {access_req.request_type}")
print(f"Deadline: {access_req.deadline.strftime('%Y-%m-%d')} (30 days from request)")
print()

# Process the request
result = handler.handle_access_request(access_req.request_id)
print("Access Request Result:")
print(json.dumps(result, indent=2, default=str))
````
