# Kernel-based regression (SVR, ROCKET)

This folder focuses on **kernel/feature-map** approaches to time-series regression:
- alignment / distance-based kernels (DTW-style) + SVR
- random convolutional feature maps (ROCKET) + linear regression

## Notebooks
- `time_series_svr_tslearn/overview.ipynb`: `TimeSeriesSVRTslearn(...)` (tslearn-inspired) with DTW-RBF option.
- `rocket_regressor/overview.ipynb`: `RocketRegressor(...)` (sktime-inspired) using ROCKET features + `RidgeCV`.
