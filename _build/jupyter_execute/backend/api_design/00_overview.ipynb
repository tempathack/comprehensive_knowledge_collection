# API Design

APIs are contracts. Good API design balances **clarity**, **stability**, **performance**, and **security**.


## REST fundamentals
- Resources, not verbs
- Consistent status codes
- Idempotency for safe retries

## Versioning
- URL versioning (`/v1`) vs header versioning
- Breaking vs non‑breaking changes


```python
# Example: FastAPI route (illustrative)
from fastapi import FastAPI
app = FastAPI()

@app.get("/v1/users/{user_id}")
def get_user(user_id: int):
    return {"id": user_id, "name": "Ada"}

```

## Pagination & filtering
- Cursor‑based pagination for stability
- Filter + sort parameters should be explicit and documented
