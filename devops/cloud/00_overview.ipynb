# Cloud - Overview

Cloud computing provides on-demand infrastructure and managed services with elastic scaling and pay-as-you-go pricing.

![Cloud computing overview](https://upload.wikimedia.org/wikipedia/commons/b/b5/Cloud_computing.svg)

## Service models
| Model | What you manage | Examples |
| --- | --- | --- |
| IaaS | OS, runtime, apps | EC2, GCE, Azure VMs |
| PaaS | Apps and data | App Engine, Heroku |
| SaaS | Just your data | Gmail, Salesforce |

## Deployment models
- Public cloud
- Private cloud
- Hybrid and multi-cloud

## Shared responsibility
Providers secure the cloud; customers secure what they run in it. This affects IAM, data encryption, and network controls.

## Mini example
Compare a simple cost profile across strategies.


```python
import pandas as pd
import plotly.express as px

cost = pd.DataFrame({
    "strategy": ["On-prem (capex)", "Cloud (opex)", "Managed service"],
    "year_1_cost_k": [350, 180, 220],
})

fig = px.bar(cost, x="strategy", y="year_1_cost_k", title="Illustrative year-1 cost")
fig.update_layout(yaxis_title="Cost (thousands)")
fig.show()
```
