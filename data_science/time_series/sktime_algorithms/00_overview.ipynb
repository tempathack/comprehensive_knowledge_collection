# sktime Algorithms - Overview

This section organizes sktime methods by task and estimator type. Use it to jump to the right family (forecasting, classification, regression, clustering, anomaly detection, segmentation, transformers).


## Orientation

- Start with **Data Containers & mtypes** to get your data shape right.
- Use **Model Selection & CV** to evaluate and tune time-aware models.
- Browse the **Catalog** and **Registry** for an exhaustive, version-aware list of estimators.
- Then dive into task-specific notebooks for forecasting, classification, regression, clustering, anomaly detection, and segmentation.
