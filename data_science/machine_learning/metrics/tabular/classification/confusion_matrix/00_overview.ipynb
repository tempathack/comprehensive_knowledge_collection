# confusion_matrix

Compute the confusion matrix.

## Quick import

```python
from sklearn.metrics import confusion_matrix
```


## Notes

- **Intuition**: counts of true vs. predicted labels (rows = true, columns = predicted in scikit-learn).
- **Binary layout**: `[[TN, FP], [FN, TP]]` when labels are `[0, 1]`.
- **Pitfalls**: label order matters; pass `labels=` if you need a specific ordering. Use `normalize=` to compare across class imbalance.

## Example

Small, self-contained example:


````python
import numpy as np
from sklearn.metrics import confusion_matrix

# Toy binary classification
y_true = np.array([1, 0, 1, 1, 0, 0, 1, 0])
y_pred = np.array([1, 0, 0, 1, 0, 1, 1, 0])

confusion_matrix(y_true, y_pred)
````
