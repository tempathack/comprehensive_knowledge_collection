# label_ranking_loss

Compute label ranking loss.

## Quick import

```python
from sklearn.metrics import label_ranking_loss
```


## Notes

- **Intuition**: fraction of label pairs that are **incorrectly ordered**; lower is better.
- **Inputs**: `y_true` is a binary indicator matrix; `y_score` contains ranking scores.
- **Pitfalls**: sensitive to ties and score calibration; requires multilabel format.

## Example

Small, self-contained example:


````python
import numpy as np
from sklearn.metrics import label_ranking_loss

# 3 samples, 3 labels
y_true = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 0],
])
y_score = np.array([
    [0.9, 0.1, 0.8],
    [0.2, 0.7, 0.1],
    [0.6, 0.4, 0.3],
])

label_ranking_loss(y_true, y_score)
````
