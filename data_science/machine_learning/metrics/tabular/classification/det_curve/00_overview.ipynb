# det_curve

Compute the DET curve (false positive rate vs false negative rate).

## Quick import

```python
from sklearn.metrics import det_curve
```


## Notes

- **Intuition**: DET curve plots **false positive rate (FPR)** vs **false negative rate (FNR)** across thresholds.
- **Outputs**: returns FPR, FNR, and thresholds for the positive class.
- **Pitfalls**: defined for binary classification and **requires scores**, not hard labels.

## Example

Small, self-contained example:


````python
import numpy as np
from sklearn.metrics import det_curve

# Scores for the positive class
y_true = np.array([0, 0, 1, 1])
y_score = np.array([0.1, 0.4, 0.35, 0.8])

fpr, fnr, thresholds = det_curve(y_true, y_score)
fpr, fnr, thresholds
````
