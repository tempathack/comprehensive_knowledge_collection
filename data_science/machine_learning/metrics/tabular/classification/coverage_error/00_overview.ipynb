# coverage_error

Compute coverage error for multilabel ranking.

## Quick import

```python
from sklearn.metrics import coverage_error
```


## Notes

- **Intuition**: average number of top-ranked labels needed to cover all true labels per sample.
- **Range**: lower is better; values are not normalized to [0, 1].
- **Pitfalls**: depends on label count and class imbalance; interpret relative to baselines.

## Example

Small, self-contained example:


````python
import numpy as np
from sklearn.metrics import coverage_error

# 3 samples, 3 labels
y_true = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 0],
])
y_score = np.array([
    [0.9, 0.1, 0.8],
    [0.2, 0.7, 0.1],
    [0.6, 0.4, 0.3],
])

coverage_error(y_true, y_score)
````
