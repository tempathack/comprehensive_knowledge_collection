# multilabel_confusion_matrix

Compute a multilabel confusion matrix.

## Quick import

```python
from sklearn.metrics import multilabel_confusion_matrix
```


## Notes

- **Intuition**: computes a 2x2 confusion matrix **per label**, treating each label as a binary task.
- **Shape**: output has shape `(n_labels, 2, 2)`.
- **Pitfalls**: inputs must be multilabel indicators with shape `(n_samples, n_labels)`.

## Example

Small, self-contained example:


````python
import numpy as np
from sklearn.metrics import multilabel_confusion_matrix

# Toy multilabel example (3 labels)
y_true = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [1, 1, 0],
])
y_pred = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [1, 0, 1],
])

multilabel_confusion_matrix(y_true, y_pred)
````
