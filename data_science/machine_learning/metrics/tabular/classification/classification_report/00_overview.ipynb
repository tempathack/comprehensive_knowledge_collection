# classification_report

Build a text report showing the main classification metrics.

## Quick import

```python
from sklearn.metrics import classification_report
```


## Notes

- **Intuition**: a compact summary of precision, recall, F1-score, and support per class.
- **Programmatic use**: pass `output_dict=True` to get a dict instead of a formatted string.
- **Pitfalls**: classes with zero predicted samples yield undefined metrics; control with `zero_division`.

## Example

Small, self-contained example:


````python
import numpy as np
from sklearn.metrics import classification_report

# Toy multiclass example
y_true = np.array([0, 1, 2, 2, 1, 0])
y_pred = np.array([0, 2, 2, 2, 1, 0])

print(classification_report(y_true, y_pred))
````
