# Ensembles (Time-Series Forest)

This folder focuses on **ensemble** methods for time-series supervised learning (panel regression/classification style).

## Notebooks
- `time_series_forest_regressor/overview.ipynb`: implement `TimeSeriesForestRegressor(min_interval=..., ...)` (TSF) + Plotly demo.
- `composable_time_series_forest_regressor/overview.ipynb`: TSF variant where you provide the interval summary feature functions.
