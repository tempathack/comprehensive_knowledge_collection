# GCP Database Services

Google Cloud Platform (GCP) offers a comprehensive suite of managed database services designed to handle various workloads, from traditional relational databases to globally distributed systems and analytics warehouses.

## Overview

| Service | Type | Best For |
|---------|------|----------|
| **Cloud SQL** | Managed Relational | Traditional OLTP workloads |
| **Cloud Spanner** | Globally Distributed Relational | Global transactions with strong consistency |
| **BigQuery** | Serverless Data Warehouse | Analytics and BI at scale |
| **Firestore** | Serverless Document DB | Mobile/web apps, real-time sync |
| **Bigtable** | Wide-Column Store | IoT, time-series, analytics |

---

## 1. Cloud SQL

**Fully managed relational database service** supporting MySQL, PostgreSQL, and SQL Server.

### Key Features

| Feature | Description |
|---------|-------------|
| **Managed Infrastructure** | Automated backups, replication, patches, and updates |
| **High Availability** | Regional instances with automatic failover |
| **Scalability** | Vertical scaling up to 96 vCPUs and 624 GB RAM |
| **Security** | Encryption at rest and in transit, VPC peering, IAM integration |
| **Read Replicas** | Up to 10 read replicas for read-heavy workloads |

## 2. Cloud Spanner

**Globally distributed, strongly consistent relational database** that combines the benefits of relational structure with horizontal scalability.

### Key Features

| Feature | Description |
|---------|-------------|
| **Global Distribution** | Data replicated across regions worldwide |
| **Strong Consistency** | External consistency (strongest form of consistency) |
| **Horizontal Scaling** | Add nodes to scale reads and writes linearly |
| **99.999% SLA** | Five 9s availability for multi-region configs |
| **SQL Support** | ANSI 2011 SQL with extensions |

## 3. BigQuery

**Serverless, highly scalable enterprise data warehouse** designed for business agility and real-time analytics.

### Key Features

| Feature | Description |
|---------|-------------|
| **Serverless** | No infrastructure to manage, automatic scaling |
| **Petabyte Scale** | Query petabytes in seconds |
| **Standard SQL** | ANSI SQL:2011 compliant |
| **ML Integration** | BigQuery ML for in-database machine learning |
| **Real-time** | Streaming ingestion for real-time analytics |
| **Separation** | Compute and storage are decoupled |
