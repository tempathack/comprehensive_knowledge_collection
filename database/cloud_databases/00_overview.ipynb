## 1. Managed vs Self-Managed Databases

When deploying databases in the cloud, you have two primary approaches:

### Managed Databases (DBaaS)

The cloud provider handles infrastructure, patching, backups, and scaling.

| Aspect | Description |
|--------|-------------|
| **Infrastructure** | Provider manages servers, storage, networking |
| **Maintenance** | Automatic patching, updates, and security fixes |
| **Backups** | Automated backups with point-in-time recovery |
| **Scaling** | One-click or automatic scaling capabilities |
| **High Availability** | Built-in replication and failover |
| **Monitoring** | Integrated metrics, logs, and alerting |

**Examples:** AWS RDS, Azure SQL Database, Google Cloud SQL, Amazon DynamoDB, Azure Cosmos DB

### Self-Managed Databases

You deploy and manage the database software on cloud VMs or containers.

| Aspect | Description |
|--------|-------------|
| **Infrastructure** | You provision and manage VMs/containers |
| **Maintenance** | Manual patching, updates, and configuration |
| **Backups** | Custom backup scripts and storage |
| **Scaling** | Manual scaling or custom automation |
| **High Availability** | Configure replication and failover yourself |
| **Monitoring** | Set up your own monitoring stack |

### Comparison Matrix

| Factor | Managed | Self-Managed |
|--------|---------|---------------|
| **Operational Overhead** | Low | High |
| **Cost** | Higher per-unit | Lower per-unit (but hidden ops cost) |
| **Customization** | Limited | Full control |
| **Time to Deploy** | Minutes | Hours to days |
| **Expertise Required** | Lower | Higher |
| **Vendor Lock-in** | Higher | Lower |

## 3. Major Cloud Provider Database Offerings

### Overview by Provider

| Category | AWS | Azure | GCP |
|----------|-----|-------|-----|
| **Relational** | RDS, Aurora | SQL Database, SQL MI | Cloud SQL, AlloyDB |
| **NoSQL Document** | DynamoDB, DocumentDB | Cosmos DB | Firestore, MongoDB Atlas |
| **Key-Value** | DynamoDB, ElastiCache | Cosmos DB, Cache for Redis | Memorystore |
| **Graph** | Neptune | Cosmos DB (Gremlin) | â€” |
| **Time Series** | Timestream | Time Series Insights | â€” |
| **Data Warehouse** | Redshift | Synapse Analytics | BigQuery |

````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd

# Cloud Provider Database Comparison Data
providers = ['AWS', 'Azure', 'GCP']

# Feature scores (1-10 scale, based on general market perception)
comparison_data = {
    'Provider': providers,
    'Relational DB Options': [9, 8, 7],
    'NoSQL Options': [9, 9, 7],
    'Serverless Support': [9, 8, 9],
    'Global Distribution': [8, 9, 8],
    'Pricing Flexibility': [7, 7, 8],
    'Enterprise Features': [9, 9, 7],
}

df = pd.DataFrame(comparison_data)
categories = list(comparison_data.keys())[1:]  # Exclude 'Provider'

# Create radar chart
fig = go.Figure()

colors = ['#FF9900', '#0089D6', '#4285F4']  # AWS orange, Azure blue, GCP blue

for i, provider in enumerate(providers):
    values = df[df['Provider'] == provider][categories].values.flatten().tolist()
    values += values[:1]  # Close the polygon
    
    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=categories + [categories[0]],
        fill='toself',
        name=provider,
        line_color=colors[i],
        opacity=0.6
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[0, 10]
        )
    ),
    showlegend=True,
    title=dict(
        text='Cloud Database Provider Comparison',
        x=0.5,
        font=dict(size=16)
    ),
    height=500,
    width=700
)

fig
````

````python
# Database Service Comparison by Category
import plotly.express as px

# Number of managed database services per category
services_data = {
    'Category': ['Relational', 'NoSQL', 'In-Memory', 'Graph', 'Time Series', 'Data Warehouse'],
    'AWS': [4, 3, 2, 1, 1, 1],
    'Azure': [3, 2, 2, 1, 1, 1],
    'GCP': [3, 2, 1, 0, 0, 1]
}

df_services = pd.DataFrame(services_data)

# Reshape for grouped bar chart
df_melted = df_services.melt(id_vars=['Category'], 
                              value_vars=['AWS', 'Azure', 'GCP'],
                              var_name='Provider', 
                              value_name='Services')

fig2 = px.bar(
    df_melted,
    x='Category',
    y='Services',
    color='Provider',
    barmode='group',
    title='Managed Database Services by Category',
    color_discrete_map={'AWS': '#FF9900', 'Azure': '#0089D6', 'GCP': '#4285F4'}
)

fig2.update_layout(
    xaxis_title='Database Category',
    yaxis_title='Number of Services',
    height=400,
    width=700
)

fig2.show()
````
