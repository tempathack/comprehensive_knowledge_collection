## NoSQL Database Types

NoSQL ("Not Only SQL") databases emerged to address limitations of traditional relational databases, particularly around **scalability**, **flexibility**, and **performance** for specific workloads.

### 1. Document Databases
Store data as semi-structured documents (JSON, BSON, XML).

| Feature | Description |
|---------|-------------|
| **Structure** | Flexible schema, nested documents |
| **Query** | Rich queries on document fields |
| **Examples** | MongoDB, CouchDB, Firestore |
| **Best For** | Content management, catalogs, user profiles |

### 2. Key-Value Stores
Simplest NoSQL model â€” data stored as key-value pairs.

| Feature | Description |
|---------|-------------|
| **Structure** | Opaque values, identified by unique keys |
| **Query** | Get/Set by key only |
| **Examples** | Redis, Amazon DynamoDB, Memcached |
| **Best For** | Caching, session storage, real-time data |

### 3. Column-Family Databases
Store data in columns rather than rows, optimized for analytical queries.

| Feature | Description |
|---------|-------------|
| **Structure** | Column families, sparse data support |
| **Query** | Efficient range scans, aggregations |
| **Examples** | Apache Cassandra, HBase, ScyllaDB |
| **Best For** | Time-series, IoT, write-heavy workloads |

### 4. Graph Databases
Designed for highly connected data with complex relationships.

| Feature | Description |
|---------|-------------|
| **Structure** | Nodes, edges, and properties |
| **Query** | Traversal-based (Cypher, Gremlin) |
| **Examples** | Neo4j, Amazon Neptune, ArangoDB |
| **Best For** | Social networks, fraud detection, recommendations |

## Comparison: NoSQL Types by Use Case

The following visualization compares how well each NoSQL type fits different use case categories.

````python
import plotly.graph_objects as go

# Define NoSQL types and use case scores (1-5 scale)
categories = ['Flexibility', 'Read Performance', 'Write Performance', 
              'Complex Queries', 'Relationships', 'Scalability']

nosql_scores = {
    'Document': [5, 4, 4, 4, 2, 4],
    'Key-Value': [2, 5, 5, 1, 1, 5],
    'Column-Family': [3, 4, 5, 3, 2, 5],
    'Graph': [3, 3, 3, 4, 5, 3]
}

colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA']

fig = go.Figure()

for idx, (db_type, scores) in enumerate(nosql_scores.items()):
    fig.add_trace(go.Scatterpolar(
        r=scores + [scores[0]],  # Close the polygon
        theta=categories + [categories[0]],
        fill='toself',
        name=db_type,
        line=dict(color=colors[idx]),
        opacity=0.6
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, 5])
    ),
    title=dict(text='NoSQL Database Types: Strengths Comparison', x=0.5),
    showlegend=True,
    template='plotly_white',
    height=500
)

fig
````

## Document Structure Example

Document databases store data as nested, flexible structures. Here's how a typical e-commerce product might be represented:

````python
import json

# Example: E-commerce product document (MongoDB-style)
product_document = {
    "_id": "prod_12345",
    "name": "Wireless Noise-Canceling Headphones",
    "brand": "AudioTech",
    "price": {
        "amount": 299.99,
        "currency": "USD"
    },
    "category": ["Electronics", "Audio", "Headphones"],
    "specifications": {
        "battery_life_hours": 30,
        "bluetooth_version": "5.2",
        "weight_grams": 250,
        "noise_cancellation": True
    },
    "inventory": {
        "in_stock": True,
        "quantity": 150,
        "warehouse_locations": ["NYC-01", "LA-03"]
    },
    "reviews": [
        {"user": "user_789", "rating": 5, "comment": "Best headphones ever!"},
        {"user": "user_456", "rating": 4, "comment": "Great sound, comfortable fit"}
    ],
    "created_at": "2025-01-15T10:30:00Z",
    "updated_at": "2026-01-28T14:22:00Z"
}

# Display the document structure
print("Product Document Structure:")
print("=" * 50)
print(json.dumps(product_document, indent=2))

# Demonstrate flexible querying
print("\n" + "=" * 50)
print("Accessing nested data:")
print(f"  Price: ${product_document['price']['amount']}")
print(f"  Battery Life: {product_document['specifications']['battery_life_hours']}h")
print(f"  Categories: {', '.join(product_document['category'])}")
print(f"  Average Rating: {sum(r['rating'] for r in product_document['reviews']) / len(product_document['reviews']):.1f}")
````

## ğŸ“Œ Key Takeaways

1. **No One-Size-Fits-All**: Each NoSQL type excels at specific use cases â€” choose based on your data access patterns, not trends.

2. **Document DBs** offer the best balance of flexibility and query power for general-purpose applications.

3. **Key-Value Stores** are unmatched for simple, high-speed lookups but sacrifice query capability.

4. **Column-Family DBs** shine in write-heavy, time-series, and analytical workloads at massive scale.

5. **Graph DBs** are essential when relationships are first-class citizens in your data model.

6. **CAP Theorem Matters**: Understand whether your application prioritizes consistency, availability, or partition tolerance.

7. **Polyglot Persistence**: Modern architectures often use multiple database types â€” relational for transactions, NoSQL for specific workloads.
