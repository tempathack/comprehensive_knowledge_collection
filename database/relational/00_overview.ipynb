## SQLite3 Example: Creating and Querying a Table

SQLite is a lightweight, file-based relational database that's perfect for learning SQL concepts.

````python
import sqlite3

# Create an in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create a table for employees
cursor.execute('''
    CREATE TABLE employees (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL,
        department TEXT NOT NULL,
        salary REAL,
        hire_date TEXT
    )
''')

# Insert sample data
employees_data = [
    ('Alice Johnson', 'Engineering', 95000, '2022-01-15'),
    ('Bob Smith', 'Marketing', 75000, '2021-06-20'),
    ('Carol Williams', 'Engineering', 105000, '2020-03-10'),
    ('David Brown', 'Sales', 80000, '2023-02-01'),
    ('Eve Davis', 'Engineering', 98000, '2021-09-05'),
]

cursor.executemany('''
    INSERT INTO employees (name, department, salary, hire_date)
    VALUES (?, ?, ?, ?)
''', employees_data)

conn.commit()

# Query: Get all employees
print("All Employees:")
print("-" * 60)
cursor.execute('SELECT * FROM employees')
for row in cursor.fetchall():
    print(f"ID: {row[0]}, Name: {row[1]}, Dept: {row[2]}, Salary: ${row[3]:,.0f}")

# Query: Aggregate - Average salary by department
print("\nAverage Salary by Department:")
print("-" * 40)
cursor.execute('''
    SELECT department, AVG(salary) as avg_salary, COUNT(*) as emp_count
    FROM employees
    GROUP BY department
    ORDER BY avg_salary DESC
''')
for row in cursor.fetchall():
    print(f"{row[0]}: ${row[1]:,.0f} (n={row[2]})")

# Clean up
conn.close()
````

## RDBMS Market Share Visualization

Let's visualize the market share of popular relational database systems.

````python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# RDBMS Market Share Data (approximate, based on DB-Engines rankings)
rdbms_data = pd.DataFrame({
    'Database': ['Oracle', 'MySQL', 'SQL Server', 'PostgreSQL', 'SQLite', 'MariaDB', 'IBM Db2'],
    'Market_Share': [28.5, 22.3, 18.7, 15.2, 6.8, 4.5, 4.0],
    'Type': ['Commercial', 'Open Source', 'Commercial', 'Open Source', 'Open Source', 'Open Source', 'Commercial'],
    'Trend': ['Stable', 'Growing', 'Stable', 'Growing Fast', 'Stable', 'Growing', 'Declining']
})

# Create a pie chart for market share
fig = px.pie(
    rdbms_data, 
    values='Market_Share', 
    names='Database',
    title='Relational Database Market Share (2025)',
    color='Type',
    color_discrete_map={'Commercial': '#1f77b4', 'Open Source': '#2ca02c'},
    hole=0.4
)

fig.update_traces(
    textposition='outside',
    textinfo='label+percent',
    hovertemplate='<b>%{label}</b><br>Market Share: %{value}%<br>Type: %{customdata}<extra></extra>',
    customdata=rdbms_data['Type']
)

fig.update_layout(
    annotations=[dict(text='RDBMS', x=0.5, y=0.5, font_size=16, showarrow=False)],
    legend_title_text='License Type',
    width=700,
    height=500
)

fig
````

````python
# Create a bar chart comparing key features
comparison_data = pd.DataFrame({
    'Feature': ['JSON Support', 'Replication', 'Full-Text Search', 'Partitioning', 'Window Functions', 'Stored Procedures'],
    'PostgreSQL': [5, 4, 5, 5, 5, 4],
    'MySQL': [4, 5, 4, 4, 4, 4],
    'SQL Server': [4, 5, 5, 5, 5, 5]
})

fig2 = go.Figure()

colors = {'PostgreSQL': '#336791', 'MySQL': '#4479A1', 'SQL Server': '#CC2927'}

for db in ['PostgreSQL', 'MySQL', 'SQL Server']:
    fig2.add_trace(go.Bar(
        name=db,
        x=comparison_data['Feature'],
        y=comparison_data[db],
        marker_color=colors[db]
    ))

fig2.update_layout(
    title='Feature Comparison: PostgreSQL vs MySQL vs SQL Server',
    xaxis_title='Feature',
    yaxis_title='Score (1-5)',
    barmode='group',
    yaxis=dict(range=[0, 5.5]),
    legend_title_text='Database',
    width=800,
    height=450
)

fig2.show()
````
