## 1. Web Fundamentals

### The Browser Rendering Pipeline

```
HTML â”€â”€> DOM Tree â”€â”€â”
                    â”œâ”€â”€> Render Tree â”€â”€> Layout â”€â”€> Paint â”€â”€> Composite
CSS â”€â”€â”€> CSSOM â”€â”€â”€â”€â”˜
```

### Critical Rendering Path

1. **Parse HTML** â†’ Build DOM (Document Object Model)
2. **Parse CSS** â†’ Build CSSOM (CSS Object Model)
3. **Execute JavaScript** â†’ Can modify DOM/CSSOM
4. **Render Tree** â†’ Combine DOM + CSSOM
5. **Layout** â†’ Calculate element positions/sizes
6. **Paint** â†’ Fill in pixels
7. **Composite** â†’ Layer and display

## 2. Frameworks & Architecture

### Framework Comparison

| Framework | Philosophy | Learning Curve | Best For |
|-----------|------------|----------------|----------|
| **React** | Library, bring your own tools | Medium | Flexible, large ecosystem |
| **Vue** | Progressive framework | Low-Medium | Quick setup, good docs |
| **Angular** | Full framework | High | Enterprise, TypeScript |
| **Svelte** | Compile-time framework | Low | Performance, simplicity |
| **Next.js** | React + SSR/SSG | Medium | Full-stack React |

### Rendering Strategies

| Strategy | Description | Pros | Cons |
|----------|-------------|------|------|
| **CSR** (Client-Side) | JS renders in browser | Interactive, simple deploy | Slow initial load, SEO issues |
| **SSR** (Server-Side) | Server renders HTML | Fast FCP, SEO friendly | Server load, complexity |
| **SSG** (Static Generation) | Build-time rendering | Fastest, CDN-friendly | Rebuild for changes |
| **ISR** (Incremental Static) | SSG + on-demand revalidation | Best of SSG + dynamic | Framework-specific |

````python
# Framework popularity and satisfaction
frameworks = ['React', 'Vue', 'Angular', 'Svelte', 'Next.js', 'Nuxt', 'SvelteKit']

usage = [82, 46, 49, 21, 60, 18, 12]  # % of developers using
satisfaction = [83, 87, 42, 96, 90, 85, 94]  # % satisfied
interest = [47, 34, 20, 70, 65, 30, 60]  # % want to learn

fig = make_subplots(rows=1, cols=2, subplot_titles=(
    'Usage vs Satisfaction', 'Framework Popularity Radar'
), specs=[[{'type': 'xy'}, {'type': 'polar'}]])

# Scatter plot
fig.add_trace(go.Scatter(
    x=usage, y=satisfaction, mode='markers+text',
    marker=dict(size=interest, sizemode='area', sizeref=0.5, color=satisfaction, colorscale='RdYlGn'),
    text=frameworks, textposition='top center',
    hovertemplate='<b>%{text}</b><br>Usage: %{x}%<br>Satisfaction: %{y}%<extra></extra>'
), row=1, col=1)

# Radar chart for top 5
categories = ['Usage', 'Satisfaction', 'Interest', 'Performance', 'DX']
react_scores = [82, 83, 47, 75, 80]
vue_scores = [46, 87, 34, 80, 90]
svelte_scores = [21, 96, 70, 95, 95]

fig.add_trace(go.Scatterpolar(
    r=react_scores + [react_scores[0]], theta=categories + [categories[0]],
    name='React', line=dict(color='#61dafb')
), row=1, col=2)
fig.add_trace(go.Scatterpolar(
    r=vue_scores + [vue_scores[0]], theta=categories + [categories[0]],
    name='Vue', line=dict(color='#42b883')
), row=1, col=2)
fig.add_trace(go.Scatterpolar(
    r=svelte_scores + [svelte_scores[0]], theta=categories + [categories[0]],
    name='Svelte', line=dict(color='#ff3e00')
), row=1, col=2)

fig.update_layout(title='Frontend Framework Landscape', template='plotly_white')
fig.update_xaxes(title_text='Usage (%)', row=1, col=1)
fig.update_yaxes(title_text='Satisfaction (%)', row=1, col=1)
fig
````

## 4. Performance

### Core Web Vitals

| Metric | Measures | Good | Needs Improvement | Poor |
|--------|----------|------|-------------------|------|
| **LCP** (Largest Contentful Paint) | Loading | â‰¤2.5s | 2.5-4s | >4s |
| **INP** (Interaction to Next Paint) | Interactivity | â‰¤200ms | 200-500ms | >500ms |
| **CLS** (Cumulative Layout Shift) | Visual stability | â‰¤0.1 | 0.1-0.25 | >0.25 |

### Performance Optimization Strategies

| Strategy | Impact | Complexity |
|----------|--------|------------|
| Code splitting | High | Medium |
| Tree shaking | High | Low |
| Image optimization | High | Low |
| Lazy loading | Medium | Low |
| Caching (Service Worker) | High | Medium |
| CDN | High | Low |
| Preloading critical assets | Medium | Low |
| Server-side rendering | High | High |

````python
# Core Web Vitals thresholds visualization
metrics = ['LCP', 'INP', 'CLS']
good_threshold = [2500, 200, 0.1]  # ms, ms, score
needs_improvement = [4000, 500, 0.25]

# Simulated website scores
websites = ['Well Optimized', 'Average Site', 'Needs Work', 'Poor']
lcp_scores = [1800, 2800, 3500, 5000]
inp_scores = [120, 250, 400, 600]
cls_scores = [0.05, 0.15, 0.22, 0.4]

fig = make_subplots(rows=1, cols=3, subplot_titles=('LCP (ms)', 'INP (ms)', 'CLS'))

# Color based on performance
def get_color(value, good, poor):
    if value <= good:
        return '#2ecc71'
    elif value <= poor:
        return '#f39c12'
    else:
        return '#e74c3c'

lcp_colors = [get_color(v, 2500, 4000) for v in lcp_scores]
inp_colors = [get_color(v, 200, 500) for v in inp_scores]
cls_colors = [get_color(v, 0.1, 0.25) for v in cls_scores]

fig.add_trace(go.Bar(x=websites, y=lcp_scores, marker_color=lcp_colors, name='LCP'), row=1, col=1)
fig.add_trace(go.Bar(x=websites, y=inp_scores, marker_color=inp_colors, name='INP'), row=1, col=2)
fig.add_trace(go.Bar(x=websites, y=cls_scores, marker_color=cls_colors, name='CLS'), row=1, col=3)

# Add threshold lines
fig.add_hline(y=2500, line_dash='dash', line_color='green', row=1, col=1)
fig.add_hline(y=4000, line_dash='dash', line_color='red', row=1, col=1)
fig.add_hline(y=200, line_dash='dash', line_color='green', row=1, col=2)
fig.add_hline(y=500, line_dash='dash', line_color='red', row=1, col=2)
fig.add_hline(y=0.1, line_dash='dash', line_color='green', row=1, col=3)
fig.add_hline(y=0.25, line_dash='dash', line_color='red', row=1, col=3)

fig.update_layout(title='Core Web Vitals by Website Performance', template='plotly_white', showlegend=False)
fig
````

## 5. Accessibility (a11y)

### WCAG Principles (POUR)

| Principle | Description | Examples |
|-----------|-------------|----------|
| **Perceivable** | Content can be perceived | Alt text, captions, contrast |
| **Operable** | UI can be operated | Keyboard nav, focus management |
| **Understandable** | Content is understandable | Clear labels, error messages |
| **Robust** | Works with assistive tech | Semantic HTML, ARIA |

### Common Accessibility Issues

| Issue | % of Pages | Fix |
|-------|------------|-----|
| Low contrast text | 83% | Ensure 4.5:1 ratio |
| Missing alt text | 58% | Add descriptive alt |
| Empty links | 50% | Add link text |
| Missing form labels | 46% | Associate labels |
| Empty buttons | 27% | Add button text |
| Missing document language | 18% | Add `lang` attribute |

````python
# Accessibility issues prevalence
issues = ['Low Contrast', 'Missing Alt Text', 'Empty Links', 
          'Missing Labels', 'Empty Buttons', 'No Lang Attr']
prevalence = [83, 58, 50, 46, 27, 18]
severity = ['High', 'High', 'Medium', 'High', 'Medium', 'Low']
severity_colors = {'High': '#e74c3c', 'Medium': '#f39c12', 'Low': '#3498db'}

fig = go.Figure()

fig.add_trace(go.Bar(
    y=issues, x=prevalence, orientation='h',
    marker_color=[severity_colors[s] for s in severity],
    text=[f'{p}%' for p in prevalence], textposition='outside'
))

fig.update_layout(
    title='Most Common Accessibility Issues (% of websites)',
    xaxis_title='Percentage of Websites with Issue',
    template='plotly_white',
    xaxis=dict(range=[0, 100])
)

# Add legend for severity
fig.add_annotation(x=85, y=5, text='<b>Severity:</b>', showarrow=False)
fig.add_annotation(x=85, y=4, text='ğŸ”´ High', showarrow=False)
fig.add_annotation(x=85, y=3, text='ğŸŸ  Medium', showarrow=False)
fig.add_annotation(x=85, y=2, text='ğŸ”µ Low', showarrow=False)

fig
````
